Para crear las tablas en Cassandra, primero debes crear un keyspace. A continuación, proporciono las queries para crear las tablas dentro del keyspace "rubenquispe" en el entorno de Astra DataStax:

# 1 Crear Keyspace:

```python
CREATE KEYSPACE IF NOT EXISTS rubenquispe
WITH replication = {'class': 'NetworkTopologyStrategy', 'replication_factor': 1};
```

# Crear Tabla1 - Pelicula_categoria:

```python
CREATE TABLE IF NOT EXISTS rubenquispe.Pelicula_categoria (
  Pelicula_Nombre text,
  Pelicula_Categoria text,
  Pelicula_Actores set<text>,
  fecha_estreno text,
  duracion text,
  PRIMARY KEY (Pelicula_Nombre, Pelicula_Categoria)
);
```

# Crear Tabla2 - Usuario_Reserva:

```python
CREATE TABLE IF NOT EXISTS rubenquispe.Usuario_Reserva (
  Usuario_DNI text,
  Reserva_Nro int,
  TipoBoleto_Nombre text,
  Reserva_Confirmado boolean,
  PRIMARY KEY (Usuario_DNI, Reserva_Nro)
);
```

# Crear Tabla3 - Cine_Sala:

```python
CREATE TABLE IF NOT EXISTS rubenquispe.Cine_Sala (
  Cine_Nombre text,
  Sala_Nro int,
  Sala_Capacidad int,
  PRIMARY KEY (Cine_Nombre, Sala_Nro)
);
```

# Crear Tabla4 - Usuario_Reservaciones:

```python
CREATE TABLE IF NOT EXISTS rubenquispe.Usuario_Reservaciones (
  Usuario_DNI text,
  Fecha date,
  num_reservas int,
  PRIMARY KEY (Usuario_DNI, Fecha)
);
```

# Crear Tabla5 - Tarjeta_Reserva:

```python
CREATE TABLE IF NOT EXISTS rubenquispe.Tarjeta_Reserva (
  Tarjeta_Banco text,
  Fecha date,
  Reserva_Nro int,
  Reserva_Confirmado boolean,
  PRIMARY KEY (Tarjeta_Banco, Fecha, Reserva_Nro)
);
```

# Crear Tabla6 - Pelicula_Categoria:

```python
CREATE TABLE IF NOT EXISTS rubenquispe.Pelicula_Categoria (
  Pelicula_Categoria text,
  Pelicula_Nombre text,
  Pelicula_Actores set<text>,
  PRIMARY KEY (Pelicula_Categoria, Pelicula_Nombre)
);
```

# Crear Tabla7 - Usuario_Telefono:

```python
CREATE TABLE IF NOT EXISTS rubenquispe.Usuario_Telefono (
  Usuario_Telefono text,
  Usuario_DNI text,
  Usuario_Nombre text,
  PRIMARY KEY (Usuario_Telefono, Usuario_DNI)
);
```

# Crear Tabla8 - Ciudad_Funcion:

```python
CREATE TABLE IF NOT EXISTS rubenquispe.Ciudad_Funcion (
  Ciudad_Nombre text,
  Funcion_FechaHora date,
  Cine_ID int,
  Cine_Nombre text,
  Sala_Nro int,
  Pelicula_Nombre text,
  Funcion_PorcentajeOcupacion float,
  PRIMARY KEY (Ciudad_Nombre, Funcion_FechaHora)
);
```

# Agregando las tablas de soporte
crear tablas de soporte para las entidades Cine y Película. Estas tablas contendrán información adicional que puede ser útil para realizar consultas de manera eficiente. Aquí las queries de creación de estas tablas de soporte:

# Tabla9 SoporteCine:

Esta tabla de soporte para la entidad Cine tiene una clave primaria compuesta por el Cine_ID. Puedes almacenar información relevante de Cine en esta tabla para facilitar las consultas.

```python
CREATE TABLE IF NOT EXISTS rubenquispe.SoporteCine (
  Cine_ID int PRIMARY KEY,
  Cine_Nombre text,
  Ciudad_Nombre text
);
```

# Tabla 10 SoportePelicula:

Esta tabla de soporte para la entidad Película tiene una clave primaria compuesta por el Pelicula_Nombre. Al igual que con la tabla SoporteCine, puedes almacenar información adicional de Película para mejorar la eficiencia de las consultas.

```python
CREATE TABLE IF NOT EXISTS rubenquispe.SoportePelicula (
  Pelicula_Nombre text PRIMARY KEY,
  Pelicula_Categoria text
);
```
